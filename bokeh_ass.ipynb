```python
#Q1. How can you create a Bokeh plot using Python code?
```

```python
#answer1-
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure, show,output_file
output_file("Bokeh Plot")
p = figure()
p.circle([1, 2, 3,4], [5, 6,7,8])
show(p)
```

```python
#Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
```

```python
#answer2-

#In Bokeh, glyphs are visual markers or symbols used to represent data on a plot.They can be added to a Bokeh plot using the glyph methods provided by the figure object.

from bokeh.plotting import figure, show,output_file
output_file("Bokeh Plot")
p = figure()
p.line([1, 2, 3], [4, 5, 6])
show(p)

```

```python
#Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
```

```python
#answer3-
from bokeh.plotting import figure, show,output_file
x = [1, 2, 3, 4, 5]
y = [3,4,34,8,34]
output_file("Bokeh Plot")
p = figure(title = "scatter plot")
p.xaxis.axis_label = "x axis"
p.yaxis.axis_label = "y axis"
p.scatter(x,y)
show(p)
```

```python
#Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?
```

```python
#answer4-
"""
A Bokeh server is a way to create interactive plots that can be updated in real-time. It allows for interactions like hovering, clicking, and zooming.
To use a Bokeh server, you need to install the Bokeh library and create a Python script containing the code for the interactive plot.
"""
from bokeh.io import curdoc
from bokeh.plotting import figure
from bokeh.server.server import Server
from tornado.ioloop import IOLoop
def modify_doc(doc):

    p = figure()
    p.circle([1, 2, 3], [4, 5, 6])
    doc.add_root(p)

# Create the Bokeh server
server = Server({'/': modify_doc}, io_loop=IOLoop(), allow_websocket_origin=["localhost:8080"])

# Start the server
server.start()

# Run the application
server.io_loop.start()
```

```python

from bokeh.plotting import figure
from bokeh.embed import components

# Create a figure object
p = figure()

# Add glyphs to the figure
p.circle([1, 2, 3], [4, 5, 6])

# Get the script and div components
script, div = components(p)



from flask import Flask, render_template

app = Flask(__name__)

@app.route('/')
def index():
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run()
```
