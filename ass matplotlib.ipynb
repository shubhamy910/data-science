```python
#answer2 - 
"""
Scatter Plot: A plot that displays individual data points as markers on a two-dimensional plane. It helps in
examining the relationship between two variables and identifying any correlation.
"""
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Scatterd Plot")
plt.show()
```

```python
#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
#Use the following data:
"""
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
"""
```

```python
#answer3- 
"""
The subplot() function in Matplotlib is used to create multiple subplots (i.e., multiple plots within a single
figure) so that we can display different plots side by side or in a grid. It helps in organizing and visualizing
data efficiently.
"""
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])

# Subplot 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Line 1")

# Subplot 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("Line 2")

# Subplot 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title("Line 3")

# Subplot 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("Line 4")

# Adjust spacing between subplots
plt.tight_layout(pad = 1.8)
plt.show()
```

```python
#answer4-
"""
A bar plot, also known as a bar chart or bar graph, is a visualization that uses rectangular bars to represent categorical 
data. The length of each bar corresponds to the value or frequency of the category it represents. Bar plots are particularly 
useful for comparing and displaying data across different categories or groups.

A bar plot is commonly used for the following purposes:

1.Comparisons: Bar plots allow for easy visual comparison of values or frequencies between different categories. The height or
length of each bar provides a clear representation of the data, making it straightforward to identify and compare magnitudes or trends.

2.Categorical Data: Bar plots are suitable for visualizing categorical data, where each bar represents a distinct category. 
They can display various types of categorical information, such as counts, frequencies, percentages, or proportions.

3.Grouped Data: Bar plots can effectively display and compare data across multiple groups or subcategories.By grouping the 
bars based on different criteria, such as time periods, demographic factors, or experimental conditions, it becomes possible to 
observe patterns, differences, or similarities across the groups.

4.Distribution: In some cases, a bar plot can be used to represent the distribution of a continuous variable by dividing it 
into discrete bins or intervals. The bars then indicate the frequency or count of data falling into each bin, providing insights 
into the shape or characteristics of the distribution.

5.Presentation and Communication: Bar plots are commonly used in presentations, reports, and publications due to their 
simplicity and effectiveness in conveying information. They provide a visually appealing way to present data and make it more 
accessible to a wider audience.
"""
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("Company Name")
plt.ylabel("Profit of Company")
plt.title("Bar Plot")
```

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
plt.ylabel("Company Name")
plt.xlabel("Profit of Company")
plt.title("Horizontal Bar Plot")
```
